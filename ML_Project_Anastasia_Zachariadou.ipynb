# <span style="color: #FF00FF;">Part I: Original data preparation and set creation</span>

## <span style="color: #800080;">Data Importing</span>

```python
import pandas as pd
df = pd.read_csv('HIGGS_8K.csv', header=None)
```

```python
df.at[0,17]=0.0
df[17]=df[17].astype(float)
```

## <span style="color: #800080;">Data Categorization</span>


The initial data can be split in 3 categories:
1) classif: Classifies an event as either signal (1) or noise (0). It correspoinds to the column 0.
2) low: Values of the 21 low-level quantities for each event. These correspond to columns 1 through 21.
3) high: Values of the 7 high-level quantites for each event. These correspond to columns 22 through 28

```python
classif=df.iloc[:,0]
low=df.iloc[:,1:22]
high=df.iloc[:,22:29]
```

## <span style="color: #800080;">Training set and test set creation</span>

Each time, 80% of the data will be utilized for the creation of the training set and a random state will be selected to ensure repeatability.

# <span style="color: #FF00FF;">Part II: Classifiers</span>

In this section, two distinct classifiers will be employed to categorize the events as either signal or background noise. For each classifier, the same process will be followed once for the low-level quantity sets and once for the high-level quantity ones.

## <span style="background-color: purple; color: white;">&nbsp;k-nn Classifier&nbsp;</span> &nbsp;(k-nearest neighbors)


```python
from sklearn.neighbors import KNeighborsClassifier
```

## <span style="color: #800080;">Scaling</span>

Data scaling is a mandatory step before employing the k-nn classifier method.

```python
from sklearn.preprocessing import StandardScaler
from pandas import DataFrame

scaler=StandardScaler()

# For the sets based on low-level quantities
low_train_sc=scaler.fit_transform(low_train)
low_test_sc=scaler.transform(low_test)

# For the sets based on high-level quantities
high_train_sc=scaler.fit_transform(high_train)
high_test_sc=scaler.transform(high_test)
```

### <span style="color: #BF40BF;">Low-level</span>

A k-nn classifier is created using the scaled low-level data. The number of neighbors (k) is set equal to 8.

```python
low_knn_classifier= KNeighborsClassifier(n_neighbors=8)
low_knn_classifier.fit(low_train_sc, low_classif_train)
```

#### <span style="color: #0000FF;">k-nn classifier predictions based on low-level quantities</span>

```python
low_knn_test_pr=low_knn_classifier.predict(low_test_sc)
low_knn_test_pr
```

#### <span style="color: #0000FF;">Actual low-level classification test set values </span>

```python
low_classif_test
```

### <span style="color: #BF40BF;">High-level</span>

Same process is repeated for the creation of a k-nn classifier using the scaled high-level.

```python
high_knn_classifier= KNeighborsClassifier(n_neighbors=8)
high_knn_classifier.fit(high_train_sc, high_classif_train)
```

#### <span style="color: #0000FF;">k-nn classifier predictions based on high-level quantities</span>

```python
high_knn_test_pr=high_knn_classifier.predict(high_test_sc)
high_knn_test_pr
```

#### <span style="color: #0000FF;">Actual high-level classification test set values </span>

```python
high_classif_test
```

## <span style="color: #800080;">Confusion Matrix and Accuracy</span>

#### <span style="color: #0000FF;">Confusion matrix</span>

# <span style="color: #FF00FF;">Part III: Artificial Neural Networks</span>

In this section, the use of artificial neural networks is employed for the purpose of classification. As before, the same process is implemented for the building of the networks based on the low-level quantities and once based on the highlevel quantities with the help of the tensorflow library. The training and test sets have already been created and scaled in parts I and II.

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Input
```

### <span style="color: #BF40BF;">Low-level</span>

## <span style="color: #800080;">ANN Building</span>

#### <span style="color: #0000FF;">ANN Initialization</span>

```python
low_ann= tf.keras.models.Sequential()
```

#### <span style="color: #0000FF;">Adding the input layer</span>

Since the number of low-level quantities is 21, the number of nodes will be 21 as well.

```python
low_ann.add(Input(shape=(21,)))
```

#### <span style="color: #0000FF;">Adding the 1st hidden layer</span>

Number of nodes is set to 25

```python
low_ann.add(tf.keras.layers.Dense(25, activation='relu'))
```

#### <span style="color: #0000FF;">Adding the 2nd hidden layer</span>

Number of nodes is set to 21

```python
low_ann.add(tf.keras.layers.Dense(21, activation='relu'))
```

#### <span style="color: #0000FF;">Adding the output layer</span>

The output layer has a single node. Since this problem involves binary classification, the sigmoid activation function is selected.

```python
low_ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## <span style="color: #800080;">ANN Training</span>

#### <span style="color: #0000FF;">ANN Compiling</span>

Because of the binary classification nature of the problem, the following are selected: 'adam' as an optimizer, 'binary_crossentropy' as a loss function and 'binary_accuracy' as a metric

```python
low_ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
```

```python
low_ann.summary()
```

#### <span style="color: #0000FF;">ANN Training on the training set</span>

The ANN is trained based on the scaled low-level training set. Batch size is set to 16 and epochs to 150

```python
low_ann.fit(low_train_sc, low_classif_train, batch_size = 16, epochs = 150)
```

## <span style="color: #800080;">Predictions</span>

#### <span style="color: #0000FF;">ANN predictions based on the low-level quantities</span>

```python
low_ann_pr = low_ann.predict(low_test_sc)
low_ann_pr
```

```python
low_ann_pr = (low_ann_pr > 0.5)
low_ann_pr
```

## <span style="color: #800080;">Results Evaluation</span>

The last step is to evaluate the model using the confusion matrix and accuracy score. Test loss will also be evaluated.

#### <span style="color: #0000FF;">Confusion Matrix</span>

```python
low_ann_cm = confusion_matrix(low_classif_test, low_ann_pr)
print(f"Confusion Matrix of ANN based on the low-level quantities: \n{low_ann_cm}")
```

```python
# Visualization
low_ann_plot = ConfusionMatrixDisplay(confusion_matrix=low_ann_cm)
low_ann_plot.plot()
plt.title("ANN based on low-level quantities confusion matrix")
plt.show()
```

#### <span style="color: #0000FF;">Accuracy Score</span>

```python
low_ann_acc=accuracy_score(low_classif_test, low_ann_pr)
print(f"Accuracy score of ANN based on the low-level quantities: {low_ann_acc}")
```

#### <span style="color: #0000FF;">Test Loss and Accuracy</span>

```python
low_ann_loss, low_ann_accuracy,=low_ann.evaluate(low_test_sc, low_classif_test)
```

```python
print(f'Test accuracy: {low_ann_accuracy}')
print(f'Test loss: {low_ann_loss}')
```

# <span style="color: #FF00FF;">Extra Part: ROC curves and AUC score</span>


As an extra form of visualization of the results, the ROC curves and AUC score are added for all the models tested in this project. As is clear both from the curves and the area under curve socres, the ANN and Random Forest models based on high quantities produce the largest number of accurate predictions.

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
knn_low_pr_prob=low_knn_classifier.predict_proba(low_test_sc)
knn_high_pr_prob=high_knn_classifier.predict_proba(high_test_sc)
trees_low_pr_prob=low_trees_classifier.predict_proba(low_test)
trees_high_pr_prob=high_trees_classifier.predict_proba(high_test)
forest_low_pr_prob=low_forest_classifier.predict_proba(low_test)
forest_high_pr_prob=high_forest_classifier.predict_proba(high_test)
ann_low_pr_prob = low_ann.predict(low_test_sc)  
ann_high_pr_prob = high_ann.predict(high_test_sc)

fpr_knn_low, tpr_knn_low, thresholds = roc_curve(low_classif_test, knn_low_pr_prob[:,1])
fpr_knn_high, tpr_knn_high, thresholds = roc_curve(high_classif_test, knn_high_pr_prob[:,1])
fpr_trees_low, tpr_trees_low, thresholds = roc_curve(low_classif_test, trees_low_pr_prob[:,1])
fpr_trees_high, tpr_trees_high, thresholds = roc_curve(high_classif_test, trees_high_pr_prob[:,1])
fpr_forest_low, tpr_forest_low, thresholds = roc_curve(low_classif_test, forest_low_pr_prob[:,1])
fpr_forest_high, tpr_forest_high, thresholds = roc_curve(high_classif_test, forest_high_pr_prob[:,1])
fpr_ann_low, tpr_ann_low, thresholds = roc_curve(low_classif_test, ann_low_pr_prob)
fpr_ann_high, tpr_ann_high, thresholds = roc_curve(high_classif_test, ann_high_pr_prob)

# Plot ROC curves
plt.figure(figsize=(10, 8))

# Plot diagonal line
plt.plot([0, 1], [0, 1], 'k-', label='y=x')

# Plot ROC for KNN
plt.plot(fpr_knn_low, tpr_knn_low, label="KNN Low", color='blue')
plt.plot(fpr_knn_high, tpr_knn_high, label="KNN High", color='blue',linestyle='--')

# Plot ROC for Decision trees
plt.plot(fpr_trees_low, tpr_trees_low, label="Decision trees Low", color='green')
plt.plot(fpr_trees_high, tpr_trees_high, label="Decision trees High", color='green', linestyle='--')

# Plot ROC for Random Forest
plt.plot(fpr_forest_low, tpr_forest_low, label="Random Forest Low", color='magenta')
plt.plot(fpr_forest_high, tpr_forest_high, label="Random Forest High", color='magenta', linestyle='--')

# Plot ROC for ANN
plt.plot(fpr_ann_low, tpr_ann_low, label="ANN Low", color='red')
plt.plot(fpr_ann_high, tpr_ann_high, label="ANN High", color='red', linestyle='--')

plt.xlabel('False Positive', fontsize=18)
plt.ylabel('True Positive', fontsize=18)
plt.legend(fontsize=15)

plt.show()
```

```python
print(f'model knn low AUC score: {roc_auc_score(low_classif_test, knn_low_pr_prob[:,1])}') 
print(f'model knn high AUC score: {roc_auc_score(high_classif_test, knn_high_pr_prob[:,1])}') 
print(f'model Decision trees low AUC score: {roc_auc_score(low_classif_test, trees_low_pr_prob[:,1])}') 
print(f'model Decision trees high AUC score: {roc_auc_score(high_classif_test, trees_high_pr_prob[:,1])}') 
print(f'model Random Forest low AUC score: {roc_auc_score(low_classif_test, forest_low_pr_prob[:,1])}') 
print(f'model Random Forest high AUC score: {roc_auc_score(high_classif_test, forest_high_pr_prob[:,1])}') 
print(f'model ANN low AUC score: {roc_auc_score(low_classif_test, ann_low_pr_prob)}')
print(f'model ANN high AUC score: {roc_auc_score(high_classif_test, ann_high_pr_prob)}')
```
